```python
# coding: utf-8

get_ipython().magic('matplotlib inline')


import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras import utils

np.random.seed(25)


# 下載 mnist 資料
(x_train, y_train), (x_test, y_test) = mnist.load_data()
# x_train: trainning data
# y_train: labels of each trainning data
# x_test: testing data
# y_test: labels of each testing data

plt.figure(figsize=(4,4))
for i in range(0, 9):
    plt.subplot(3, 3, i+1)
    plt.imshow(x_train[i], cmap="gray")
    plt.title("Class %s" % y_train[i])


# add channel  開始為 CNN 準備  矩陣要變成 (資料筆數, 長, 寬, 黑白=1/彩色=3)
x_train = np.expand_dims(x_train, axis=3)
x_test = np.expand_dims(x_test, axis=3)

## 手動將 mnist 的訓練資料依 80%/20% 的比例分做訓練及驗證資料
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(
    x_train, y_train, test_size=0.2, random_state=25
)

print(x_train.shape)
print(x_test.shape)

x_train_normalized = x_train / 255
x_val_normalized = x_val / 255
x_test_normalized = x_test / 255

# RNN 輸入改成 (資料筆數, 28, 28)
x_val_rnn = x_val_normalized.reshape(-1, 28, 28)
x_test_rnn = x_test_normalized.reshape(-1, 28, 28)

y_train_onehot = utils.to_categorical(y_train)  ## 轉成 10個0/1 碼
y_val_onehot = utils.to_categorical(y_val)
y_test_onehot = utils.to_categorical(y_test)


def show_train_history(train_history, train, validation):
    plt.plot(train_history.history[train])
    plt.plot(train_history.history[validation])
    plt.title("Train History")
    plt.ylabel(train)
    plt.xlabel('Epoch')
    plt.show()


from keras.utils import *
from keras.models import Sequential
from keras.layers import Dense, Dropout, SimpleRNN
from keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator


model = Sequential()
model.add(SimpleRNN(128, return_sequences=True, input_shape=(28, 28)))
model.add(Dropout(0.3))
model.add(SimpleRNN(64, return_sequences=True))
model.add(Dropout(0.3))
model.add(SimpleRNN(32))
model.add(Dropout(0.3))
model.add(Dense(128, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(10, activation="softmax"))

model.summary()


# 定義訓練方式
# loss function: 交叉熵
# optimizer: Adam
# 評估模型: 準確率
model.compile(loss='categorical_crossentropy', optimizer=Adam(), metrics=['accuracy'])


## 訓練資料開始做一些調整轉換

## rotation_range: 定義影像旋轉的角度區間，generator 會亂數來決定選轉角度
## width_shift_range: 定義影像橫向偏移位置多寡，generator 會亂數決定偏移量
## height_shift_range: 定義影像縱向偏移位置多寡，generator 會亂數決定偏移量
## shear_range: 定義錯切角度
## zoom_range: 定義放大縮小多寡

gen = ImageDataGenerator(
    featurewise_center=False,
    samplewise_center=False,
    rotation_range=8,
    width_shift_range=0.08,
    shear_range=0.3,
    height_shift_range=0.08,
    zoom_range=0.08,
    data_format="channels_last"
)

## 原始影像增強 generator
train_generator = gen.flow(x_train_normalized, y_train_onehot, batch_size=30)

## 包一層給 RNN 用：把 (batch, 28, 28, 1) -> (batch, 28, 28)
def rnn_generator(image_generator):
    while True:
        batch_x, batch_y = next(image_generator)
        batch_x = batch_x.reshape(batch_x.shape[0], 28, 28)
        yield batch_x, batch_y

train_rnn_generator = rnn_generator(train_generator)


from keras.callbacks import ReduceLROnPlateau

learning_rate_function = ReduceLROnPlateau(
    monitor='val_accuracy',
    patience=3,
    verbose=1,
    factor=0.5,
    min_lr=0.00001
)


# 開始訓練
train_history = model.fit(
    train_rnn_generator,
    steps_per_epoch=len(x_train_normalized) // 30,
    epochs=10,
    verbose=1,
    validation_data=(x_val_rnn, y_val_onehot),
    callbacks=[learning_rate_function]
)


show_train_history(train_history, 'loss', 'val_loss')


score = model.evaluate(x_test_rnn, y_test_onehot)
print()
print("Accuracy {}%".format(score[1]))

results = model.predict(x_test_rnn)
print()
print(results)
```

```python

import itertools
def plot_confusion_matrix(cm, classes, normalize=False, title="Confusion Matrix", cmap=plt.cm.Blues):
    plt.figure()
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    thresh = cm.max() / 2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(
            j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black"
        )

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')


results = np.argmax(model.predict(x_test_rnn), axis=-1)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, results)

plot_confusion_matrix(cm, range(0, 10))

```
